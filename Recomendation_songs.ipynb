```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
features = ['id','eng_track_title','lang','album_title', 'lyrics']
```

```python
def combine_features(row):
    return row['eng_track_title']+" "+row['lang']+" "+row['album_title']+" "+row['lyrics']
```

```python
for feature in features:
    df[feature] = df[feature].fillna('')
```

```python
df["combined_features"] = df.apply(combine_features,axis=1) 
```

```python
df.head(50)
```

```python
cv = CountVectorizer()
count_matrix = cv.fit_transform(df["combined_features"])
```

```python
print(count_matrix.toarray())
```

```python
cosine_sim = cosine_similarity(count_matrix)
cosine_sim 
```

```python
def get_title_from_index(id):
    return df[df.id == id]["eng_track_title"].values[0]  
def get_index_from_title(eng_track_title):
    return df[df.eng_track_title == eng_track_title]["id"].values[0]
```

```python
song_selected = "Butter"
song_id = get_index_from_title(song_selected)

print(song_id)
similar_songs = list(enumerate(cosine_sim[song_id]))
```

```python
similar_songs
```

```python
sorted_similar_songs = sorted(similar_songs,key=lambda x:x[1],reverse=True)[1:]
sorted_similar_songs
```

```python
i=0
print("Top 5 similar songs to "+song_selected+" are:\n")
for element in sorted_similar_songs:
    #print(element)
    print(get_title_from_index(element[0]))
    i=i+1
    if i>5:
        break
```
